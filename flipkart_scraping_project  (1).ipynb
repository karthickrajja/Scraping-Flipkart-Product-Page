# Using requests library download the webpage

```python
!pip install requests
import requests as rq
```

### Landing page link of Keyword "Mobile" has been taken as input to download

```python
website_link= ("https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off")
website_data = rq.get(website_link)

data_text = website_data.text

data_text
```

# Using beautifulsoup library Parse HTML code

```python
!pip install beautifulSoup4
from bs4 import BeautifulSoup as soup
```

```python
website = soup(data_text,"html.parser")
```

# Extracting data from website page

### Product_title

```python
ttag = "_4rR01T"

product_title = website.find_all("div", {"class":ttag})

title = []
for i in product_title:
    title.append(i.text)
    
title 
```

### Star_rating

```python
rtag = "_1lRcqv"

Starrating = website.find_all("span", {"class":rtag})

Star_rating = []

for i in Starrating:
    Star_rating.append(i.text)
    
Star_rating
```

# Number of ratings and reviews

```python
ratingtage = "_2_R_DZ"

ratingtag = website.find_all("span", {"class":ratingtage})

rating = []
for i in ratingtag:
    rating.append(i.text)
rating
```

# Product_price

```python
pricetag = "_30jeq3 _1_WHN1"
pricetags = website.find_all("div", {"class":pricetag})

price = []
for i in pricetags:
    price.append(i.text)
price
len(price)
```

# Creating DataFrame using extracted data

### Using Pandas library creating dataframe

```python
import pandas as pd
```

```python
dic = {"Product_title": title ,"Price": price , "Star rating": Star_rating, "Number of Review & rating": rating }
```

```python
flipkart_page = pd.DataFrame(dic)
```

# Exporting data as CSV file

```python
flipkart_page.to_csv("flipkart_page.csv")
```
